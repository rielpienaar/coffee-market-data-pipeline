```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


green_prices_raw = pd.read_csv('green coffee prices.csv')
production_raw = pd.read_csv('Coffee_production.csv')
coffee_export_raw = pd.read_csv('Coffee_export.csv')
coffee_import_raw = pd.read_csv('Coffee_import.csv')
consumption_importers_raw = pd.read_csv('Coffee_importers_consumption.csv')
consumption_domestic_raw = pd.read_csv('Coffee_domestic_consumption.csv')

# work copies
green_prices = green_prices_raw.copy()
production = production_raw.copy()
coffee_export = coffee_export_raw.copy()
coffee_import = coffee_import_raw.copy()
consumption_importers = consumption_importers_raw.copy()
consumption_domestic = consumption_domestic_raw.copy()

display(green_prices.head(2))
display(production.head(2))
display(coffee_export.head(2))    
display(coffee_import.head(2))
display(consumption_importers.head(2))
display(consumption_domestic.head(2))
```

```python
# quick profile check for each dataframe

dfs = {
    "green_prices": green_prices,
    "production": production,
    "coffee_export": coffee_export,
    "coffee_import": coffee_import,
    "consumption_importers": consumption_importers,
    "consumption_domestic": consumption_domestic,
}

for name, df in dfs.items():
    print("\n" + "="*80)
    print(name, df.shape)
    print(df.info())
    print("Missing:\n", df.isna().sum().sort_values(ascending=False).head(15))
    print("Duplicates:", df.duplicated().sum())
```

```python
# standardize column names
# strips whitespace, converts to lowercase, replaces dots and non-alphanumeric characters with underscores, and removes leading/trailing underscores
for df in [
    green_prices,
    production,
    coffee_export,
    coffee_import,
    consumption_importers,
    consumption_domestic,
]:
    df.columns = (
        df.columns
          .str.strip()
          .str.lower()
          .str.replace(".", "_", regex=False)
          .str.replace(r"[^\w]+", "_", regex=True)
          .str.replace(r"^(\d{4})_\d{2}$", r"\1", regex=True)
          .str.strip("_")
    )

# check all dataframes again after column name standardization
for name, df in dfs.items():
    print("\n" + "="*80)
    print(name, df.shape)
    print(df.columns)
```

```python
# remove only the last columns before pivoting
production = production.iloc[:, :-1]
coffee_export = coffee_export.iloc[:, :-1]
coffee_import = coffee_import.iloc[:, :-1]
consumption_importers = consumption_importers.iloc[:, :-1]
consumption_domestic = consumption_domestic.iloc[:, :-1]


# check column names again after removing total rows to make sure we don't have any duplicates or issues before pivoting
print("Production columns:", production.columns.tolist())
print("Export columns:", coffee_export.columns.tolist())
print("Import columns:", coffee_import.columns.tolist())
print("Consumption Importers columns:", consumption_importers.columns.tolist())
print("Consumption Domestic columns:", consumption_domestic.columns.tolist())
                                                                     
```

```python
# pivot the production, export, import, and consumption dataframes to have years as rows and countries as columns
# ensure columns are country, coffee_type (only in production), year, amount for easier analysis and merging later on
production_pivot = production.melt(id_vars=["country", "coffee_type"], var_name="year", value_name="production")
export_pivot = coffee_export.melt(id_vars=["country"], var_name="year", value_name="export")
import_pivot = coffee_import.melt(id_vars=["country"], var_name="year", value_name="import")
consumption_importers_pivot = consumption_importers.melt(id_vars=["country"], var_name="year", value_name="consumption_importers")
consumption_domestic_pivot = consumption_domestic.melt(id_vars=["country","coffee_type"], var_name="year", value_name="consumption_domestic")

# convert year to numeric as integer, coercing errors to NaN (e.g. if there are any non-year values in the year column)
for df in [
    production_pivot,
    export_pivot,
    import_pivot,
    consumption_importers_pivot,
    consumption_domestic_pivot,
]:
    df["year"] = pd.to_numeric(df["year"], errors="coerce").astype("Int64")

# check the pivoted dataframes
display(production_pivot.head(2))
display(export_pivot.head(2))
display(import_pivot.head(2))
display(consumption_importers_pivot.head(2))
display(consumption_domestic_pivot.head(2))

```

```python
# standardize country names in all dataframes to ensure consistency for merging and analysis
# this is a simple standardization that can be expanded with a mapping dictionary for known variations (e.g. "United States" vs "USA")
for df in [
    production_pivot,
    export_pivot,
    import_pivot,
    consumption_importers_pivot,
    consumption_domestic_pivot,
]:
    df["country"] = (
        df["country"]
          .str.strip()
          .str.title()
          .str.replace(r"\s+", " ", regex=True)  # replace multiple spaces with single space
          .str.replace(r"[^\w\s]", "", regex=True)  # remove punctuation
    )

#check the pivot dataframes again after country name standardization
for df in [
    production_pivot,
    export_pivot,
    import_pivot,
    consumption_importers_pivot,
    consumption_domestic_pivot,
]:
    print(df["country"].value_counts())
```

```python
# correct country names based on the unique values observed in the previous step, using a mapping dictionary to replace known variations with a standardized name
country_map = {
    "Bolivia Plurinational State Of": "Bolivia",
    "Timorleste": "Timor-Leste",
    "Côte Divoire": "Côte d’Ivoire",
    "Democratic Republic Of Congo": "Democratic Republic of the Congo",
    "Congo": "Republic of the Congo",
    "Lao Peoples Democratic Republic": "Lao People’s Democratic Republic",
    "Trinidad Tobago": "Trinidad and Tobago",
    "Belgiumluxembourg": "Belgium-Luxembourg"
    }

for df in [
    production_pivot,
    export_pivot,
    import_pivot,
    consumption_importers_pivot,
    consumption_domestic_pivot,
]:
    df["country"] = df["country"].astype(str).str.strip().replace(country_map)
```

```python
# identify what columns in all the dataframes contain the value -2147483648 (which is the minimum value for a 32-bit integer and may indicate an error or placeholder value in the data)
for name, df in dfs.items():
    #print(df["country"].value_counts())
    print("Columns with -2147483648 value:")
    for col in df.columns:
        if (df[col] == -2147483648).any():
            print(f" - {col}")


```

```python
# looks like only 3 instances of these placeholder values in export - replace with 0
export_pivot["export"] = export_pivot["export"].replace(-2147483648, 0) 

# count after replacement to confirm that the placeholder values have been replaced
print("Count of -2147483648 in export after replacement:", (export_pivot["export"] == -2147483648).sum())
```

```python
# check if export is larger than production for each production country-year combination
merged = production_pivot.merge(export_pivot, on=["country","year"], how="left", suffixes=("_prod","_exp"))
merged["% diff"] = (merged["export"] - merged["production"]) / merged["production"] * 100
merged["export_larger_than_production"] = merged["export"] > merged["production"]
export_larger = merged[merged["export_larger_than_production"] == True]
print("Rows where export is larger than production:\n", export_larger[["country","year","production","export","% diff" ]])


# not going to drop these rows since we don't know if they are errors or if there are valid reasons for export being larger than production (e.g. re-exports, stockpiles, etc.) 
# will need to investigate further during analysis and potentially handle on a case-by-case basis depending on the country and year
```

```python
from sqlalchemy import create_engine, text

DB_USER = "***"
DB_PASS = "***"
DB_HOST = "***"
DB_PORT = "***"
DB_NAME = "coffee_market"

# connect to MySQL server
engine = create_engine(
    f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}"
)

# drop table if exists
# create new table
with engine.connect() as conn:
    conn.execute(text(f"DROP DATABASE IF EXISTS {DB_NAME}"))
    conn.execute(text(f"CREATE DATABASE {DB_NAME}"))
    print(f"Database {DB_NAME} created")

# connect to database
engine = create_engine(
    f"mysql+pymysql://{DB_USER}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)

# dataframes to load as tables into MySQL
tables = {
    "production": production_pivot,
    "export": export_pivot,
    "import": import_pivot,
    "consumption_importers": consumption_importers_pivot,
    "consumption_domestic": consumption_domestic_pivot,
    "prices": green_prices
}

# load in the dataframes
for name, df in tables.items():
    df.to_sql(
        name=name,
        con=engine,
        if_exists="replace",
        index=False,
        method="multi",
        chunksize=1000
    )
    print(f"Loaded {name} ({len(df)} rows)")

print("All tables loaded successfully")
```
